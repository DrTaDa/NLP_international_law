```python
import json
import glob
import numpy
import matplotlib.pyplot as plt

from forums import forums
```

```python
def format_forum(forum):
    forum_tokenized = forum.lower().split(" ")
    forum_tokenized_no_stopwords = []
    for word in forum_tokenized:
        if word != "and" and word != "on" and word != "of":
            forum_tokenized_no_stopwords.append(word.replace('-', ''))
    return forum_tokenized_no_stopwords
```

```python
forums_to_formatted = {}
for main_forum in forums:
    
    forums_to_formatted[main_forum] = format_forum(main_forum)
    
    for forum in forums[main_forum]:
        forums_to_formatted[forum] = format_forum(forum)

print(forums_to_formatted)
```

```python
keywords = [
    "non-intervention",
    "non-interference",
    "interfere in matters within domestic jurisdictions",
    "interfere in the domestic affairs",
    "interfere in the internal affairs",
    "intervene in matters within domestic jurisdictions",
    "intervene in the domestic affairs",
    "intervene in the internal affairs"
]
```

```python
forums_count = {}

for keyword in keywords:

    forums_count[keyword] = {}
    for k in forums.keys():
        forums_count[keyword][k] = 0
        for v in forums[k]:
            forums_count[keyword][v] = 0

    for path in glob.glob(f"./{keyword}/*.json"):
        
        print(path)
        metadata = json.load(open(path))

        got_hit = False
        
        if not "content_tokenized" in metadata:
            continue

        for forum_group in list(forums.keys())[::-1]:
            
            for forum in forums[forum_group]:
                
                for index, w in enumerate(metadata["content_tokenized"][:50]):
                    
                    formatted_forum = forums_to_formatted[forum]
                    
                    if w == formatted_forum[0]:

                        if metadata["content_tokenized"][index:index+len(formatted_forum)] == formatted_forum:
                            
                            forums_count[keyword][forum] += 1
                            got_hit = True
                            break

                if got_hit:
                    break

            if got_hit:
                break
```

```python
 print(forums_count)
```

```python
forum_group_count = {}
for keyword in forums_count:
    forum_group_count[keyword] = {}
    for forum_group in forums:
        forum_group_count[keyword][forum_group] = sum(forums_count[keyword][forum] for forum in forums[forum_group])
    
print(forum_group_count)
```

```python
for idx_keyword, keyword in enumerate(keywords):
    
    fig, ax = plt.subplots(figsize=(5, 6))

    width = 1
    xticks = []
    xticks_labels = []

    xpos = 1
    for i, forum_group in enumerate(forum_group_count[keyword]):

        ax.bar(
            xpos + 0.5 * len(forums[forum_group]) - 0.5, 
            height=forum_group_count[keyword][forum_group],
            width=len(forums[forum_group]) * width,
            zorder=1,
            alpha=0.2,
            color=f"C{i}"
        )
        
        for forum in forums[forum_group]:
            
            ax.bar(
                xpos,
                height=forums_count[keyword][forum],
                width=width,
                alpha=0.9,
                color=f"C{i}"
            )
            
            xticks.append(xpos)
            xticks_labels.append(forum)
            
            xpos += 1
            
    x = list(range(len(forums_count[keyword])))
    height = forums_count[keyword].values()

    ax.set_ylabel('Count')
    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks_labels, rotation=90)
    ax.set_title(keyword)
    ax.legend()

    fig.tight_layout()

    plt.savefig(f"count_forums_{keyword}.pdf", dpi=300)
    
    plt.show()
    plt.clf()
```

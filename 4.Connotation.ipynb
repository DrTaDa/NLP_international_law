## Non-intervention on World Map

```python
import json
import matplotlib.pyplot as plt
import glob
```

```python
country_list = json.load(open("countries.json", "r"))
```

```python
keyword = "non-intervention"
```

```python
paths = glob.glob("./{}/*.json".format(keyword))
for k, path in enumerate(paths):
    
    #print(k, len(paths))
    
    metadata = json.load(open(path))
    
    idx_noninterventation = []
    
    for idx, w in enumerate(metadata["content_tokenized"]):
        
        if w == "nonintervention":
            idx_noninterventation.append(idx)
        
        elif w == "non" and metadata["content_tokenized"][idx+1] == "intervention":
            idx_noninterventation.append(idx)
    
    if not idx_noninterventation:
        continue

    for idx in idx_noninterventation:

        for country in country_list:

            idx_start = max([0, idx-300])
            
            names_to_search = [country["en_short_name"].lower()]
            if "en_short_name1" in country:
                names_to_search.append(country["en_short_name1"].lower())
            if "en_short_name2" in country:
                names_to_search.append(country["en_short_name2"].lower())
            
            for name in names_to_search:

                splitted_name = name.split(" ")

                if len(splitted_name) > 1:

                    idx_first_part = [l for l,m in enumerate(metadata["content_tokenized"][idx_start:idx]) if m ==splitted_name[0]]

                    for idxx in idx_first_part:
                        for idx_part, parts in enumerate(splitted_name[1:]):
                            if idx_start+idxx+idx_part < len(metadata["content_tokenized"]):
                                if metadata["content_tokenized"][idx_start+idxx+idx_part+1] != parts:
                                    break
                        else:
                            print(path, splitted_name, metadata["content_tokenized"][idx_start+idxx:idx_start+idxx+5])
                            if "nonintervention_count" in country:
                                country["nonintervention_count"] += 1
                            else:
                                country["nonintervention_count"] = 1
                                metadata["valid"] = True

                else:

                    if name in metadata["content_tokenized"][idx_start:idx]:
                        if "nonintervention_count" in country:
                            country["nonintervention_count"] += 1
                        else:
                            country["nonintervention_count"] = 1
                            metadata["valid"] = True
```

```python
for country in country_list:
    if "nonintervention_count" in country:
        print(country["en_short_name"], country["nonintervention_count"])
    
count_max = max([country["nonintervention_count"] for country in country_list if "nonintervention_count" in country])
print(count_max)
valide_docs = [country["nonintervention_count"] for country in country_list if "nonintervention_count" in country]
len(valide_docs)
```

```python
import csv
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
    

cmap = mpl.cm.Blues

countries_shp = shpreader.natural_earth(
    resolution='110m',
    category='cultural',
    name='admin_0_countries'
)

ax = plt.axes(projection=ccrs.Robinson())


label = 0
for country in country_list:
    if "nonintervention_count" in country and "num_code" in country:
        
        for country_map in shpreader.Reader(countries_shp).records():
            
            if country_map.attributes["ISO_N3"].lower() == country["num_code"].lower():
            
                ax.add_geometries(
                    [country_map.geometry],
                    ccrs.PlateCarree(),
                    facecolor=cmap(np.log(1+country["nonintervention_count"]) / np.log(1+count_max), 1),
                    label=country["nonintervention_count"]
                )

plt.show()
```
